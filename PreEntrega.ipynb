### 4Ô∏è‚É£ Limpieza de datos

```python
# TODO: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados.
# Sugerencia: aplicar .drop_duplicates(), .str.strip(), .str.replace() seg√∫n el caso.
# Documentar el proceso y los resultados.
```

```python
clean_ventas = ventas.copy()
clean_marketing = marketing.copy()
clean_clientes = clientes.copy()

clean_ventas.drop_duplicates()
clean_marketing.drop_duplicates()
clean_clientes.drop_duplicates()
```

```python
def normalizarString(df):
  for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str).str.strip().str.replace(r"[\u200b\t\r\n]", "", regex=True).str.replace(" +", " ", regex=True).str.title()
  return df

clean_ventas = normalizarString(clean_ventas)
clean_marketing = normalizarString(clean_marketing)
clean_clientes = normalizarString(clean_clientes)
```

```python
def normalizarFecha(df):
  for col in df.columns:
    if "fecha" in col.lower():
      df[col] = pd.to_datetime(df[col], format='%d/%m/%Y', errors='coerce')
  return df

clean_ventas = normalizarFecha(clean_ventas)
clean_marketing = normalizarFecha(clean_marketing)
clean_clientes = normalizarFecha(clean_clientes)
```

```python
print(clean_ventas.dtypes)
print(clean_clientes.dtypes)
print(clean_marketing.dtypes)
```

```python
clean_ventas.dtypes.head()
clean_clientes.head()
clean_marketing.head()
```

```python
if "precio" in clean_ventas.columns:
    clean_ventas["precio"] = (
        clean_ventas["precio"]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    clean_ventas["precio"] = pd.to_numeric(clean_ventas["precio"], errors="coerce")
```

### 5Ô∏è‚É£ Transformaci√≥n de datos

```python
# TODO: Aplicar filtros y transformaciones para crear una tabla de ventas
# que muestre solo los productos con alto rendimiento. calcular el percentil 80
# y filtrar los productos que superen ese umbral en ventas.
#quantile(0.8)
# Sugerencia: usar .query() o condiciones con operadores l√≥gicos.
```

```python
clean_ventas['rendimiento']=clean_ventas['precio']*clean_ventas['cantidad']
```

```python
clean_ventas.head(5)
```

### 6Ô∏è‚É£ Agregaci√≥n

```python
# TODO: Resumir las ventas por categor√≠a de producto y analizar los ingresos generados.
# Sugerencia: usar .groupby() y .agg() para generar m√©tricas como suma y promedio.
```

```python
clean_ventas_agg=clean_ventas.groupby('producto').agg(
    ingreso_total=('rendimiento', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('rendimiento', 'size')
    ).reset_index()
clean_ventas_agg.head(5)
```

```python
ventas_top = clean_ventas_agg[clean_ventas_agg['ingreso_total']>clean_ventas_agg['ingreso_total'].quantile(0.8)].reset_index(drop=True).sort_values(['ingreso_total','unidades'], ascending=False)
display(ventas_top.head(20))
```

```python
# TODO: Combinar los sets de datos de ventas y marketing para obtener una visi√≥n m√°s amplia de las tendencias.
# Sugerencia: usar pd.merge() especificando la clave com√∫n entre ambos DataFrames.
# Documentar cualquier observaci√≥n relevante sobre la combinaci√≥n de datos.
df_ventas_marketing = pd.merge(clean_ventas,clean_marketing,on='producto',how='inner')
df_ventas_marketing.head(5)
```
